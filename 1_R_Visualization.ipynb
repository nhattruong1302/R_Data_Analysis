```r
# Create a ggplot
# The relationship between flipper lengths and body masses
ggplot(
  data = penguins,
  mapping = aes(
    x = flipper_length_mm,
    y = body_mass_g)
    ) +
  geom_point(
    mapping = aes(
      color = species,
      shape = species)
      ) +
  geom_smooth(method = 'lm') +

  labs(
    title = "Body mass and flipper length",
    subtitle = "Dimensions for Adelie, Chinstrap, and Gento Penguins",
    x = "Flipper length (mm)", y = "Body mass (g)",
    color = "Species", shape = "Species"
    ) +
  scale_color_colorblind()
```

### Visualize Distribution

i) Categorical variable

```r
# sort using fct_infreq
ggplot(penguins, aes(x=fct_infreq(species)))+
  geom_bar()
```

ii) Numerical variable

```r
# Histogram
ggplot(penguins, aes(x=body_mass_g))+
  geom_histogram(binwidth = 200)
```

```r
# Density
ggplot(penguins, aes(x=body_mass_g))+
  geom_density()
```

```r
# set the border line color
ggplot(penguins, aes(x = species)) +
  geom_bar(color = "blue")

# set the bar color
ggplot(penguins, aes(x = species)) +
  geom_bar(fill = "blue")
```

## Visualize relationships

i) Numerical and categorical variables

```r
ggplot(penguins, aes(x=species, y=body_mass_g)) +
  geom_boxplot()
```

```r
ggplot(penguins, aes(x = body_mass_g, color = species)) +
  geom_density(linewidth=0.75)
```

```r
ggplot(penguins, aes(x = body_mass_g, color = species, fill=species)) +
  geom_density(alpha=0.5, linewidth=0.75)
```

ii) Two categorical variable

```r
ggplot(penguins, aes(x=island, color=species, fill=species))+
  geom_bar()
```

```r
ggplot(penguins, aes(x=island, color=species, fill=species))+
  geom_bar(position="fill")
```

iii) Two numerical variables

```r
ggplot(penguins, aes(x = flipper_length_mm, y=body_mass_g))+
  geom_point()+
  geom_smooth(method='lm')
```

### iv) Three or more variables

```r
ggplot(penguins, aes(x = flipper_length_mm, y = body_mass_g)) +
  geom_point(aes(color = species, shape = island))
```

```r
ggplot(penguins, aes(x = flipper_length_mm, y = body_mass_g)) +
  geom_point(aes(color = species, shape = island)) +
  facet_wrap(~island)
```

```r
ggplot(
  data = penguins,
  mapping = aes(
    x = bill_length_mm, y = bill_depth_mm,
    color = species, shape = species
  )
) +
  geom_point() +
  labs(color = "Species")
```
